```python
import pandas as pd
import geopandas as gpd
```

```python
ice = pd.read_csv('ice_encounters_cleaned.csv')

dissolved = gpd.read_file('aor_shapefile_dissolved.shp')
```

```python
merged = dissolved.merge(ice, on="aor", how="left")
```

```python
merged.crs = "EPSG:4326"
merged.to_file("aor_encounters_merged.gpkg", driver="GPKG")
```
